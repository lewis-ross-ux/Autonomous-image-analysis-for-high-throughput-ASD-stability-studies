```python
import pandas as pd
import numpy as np
```

```python
##--- import data ---##
stability_df = pd.read_csv(r'/home/lero/idrive/cmac/DDMAP/stability_studies_master/API_stability_APR_26.csv')
descriptors_df = pd.read_csv(r'/home/lero/idrive/cmac/DDMAP/Stability_prediction_models/Smiles_calculator/api_descriptors.csv')
polymer_descriptors_df = pd.read_csv(r'/home/lero/idrive/cmac/DDMAP/Stability_prediction_models/Smiles_calculator/polymer_descriptors.csv')
```

```python
##--- merge API desc with stability data ---##
merged_df = stability_df.merge(descriptors_df, on="API", how="left")
```

```python
##--- add polymer details to samples ---##
conditions =  [
    (merged_df['well'].between(0,7)),
    (merged_df['well'].between(8,15)),
    (merged_df['well'].between(16,23)),
    (merged_df['well'].between(24,31)),
    (merged_df['well'].between(32,39)),
    (merged_df['well'].between(40,47))
]

choices = [
    'Soluplus',
    'PVP_K30',
    'Plasdone_S_630',
    'Affinisol_15LV',
    'HPMCAS_HG',
    'Pure'
]

merged_df['Polymer'] = np.select(conditions, choices, default=None)
cols=merged_df.columns.tolist()
cols.remove('Polymer')
cols.insert(3,'Polymer')
merged_df = merged_df[cols]
print('done')
```

```python
##--- merge polymer desc with stability data ---##
merged_df = merged_df.merge(polymer_descriptors_df, on="Polymer", how="left")
merged_df.to_csv(r'/home/lero/idrive/cmac/DDMAP/stability_studies_master/stability_dataset.csv', index=False)
print('done')
```
