```python
import pandas as pd
import os
```

```python
def average_calc(a, b):
    return ((a+b)/2)
```

```python
file_path=r'/home/lero/idrive/cmac/DDMAP/stability_studies_master'

folders = ['40_C_75_RH', 
           '30_C_30_RH',
           '40_C_0_RH'
           ]
```

```python
for api in folders:
    folder_path=os.path.join(file_path, api)
    for api_subfolder in os.listdir(folder_path):
        api_results = os.path.join(folder_path, api_subfolder, 'new_results')
        stability_results=os.path.join(api_results, 'stability_results.csv')
        if not os.path.exists(stability_results):
            print(f"⚠️ Missing file: {stability_results}")  
            continue  
        
        if os.path.exists(stability_results):
            df=pd.read_csv(stability_results)

            duplicate_1=df.iloc[:48].reset_index(drop=True)
            duplicate_2=df.iloc[48:].reset_index(drop=True)
            duplicate_2 = duplicate_2.rename(columns={'well': 'Duplicate', 
                                                      'unstable_timestamp':'unstable_timestamp (duplicate)', 
                                                      'status':'status (duplicate)',
                                                      'time_to_cryst': 'time_to_cryst (duplicate)'
                                                      })

            stability_summary = pd.concat([duplicate_1, duplicate_2], axis=1)
            stability_summary['time_to_cryst'] = stability_summary['time_to_cryst'].fillna(90)
            stability_summary['time_to_cryst (duplicate)'] = stability_summary['time_to_cryst (duplicate)'].fillna(90)
            stability_summary['Average days stable'] = average_calc(stability_summary['time_to_cryst'], stability_summary['time_to_cryst (duplicate)']).round(3)
            
            
            output_path=os.path.join(api_results, 'stability_summary.csv')
            
            stability_summary.to_csv(output_path, index=False)
            
            print(f'processed {stability_results} -> {output_path}')
```
