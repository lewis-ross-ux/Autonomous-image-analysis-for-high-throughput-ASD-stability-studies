```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'/home/lero/idrive/cmac/DDMAP/stability_studies_master/stability_dataset.csv', na_values='nan')
df.columns
```

```python
##--- adding the minimum days  stable column ---##
df['days_stable_min'] = df[['time_to_cryst', 'time_to_cryst (duplicate)']].min(axis=1)

cols = df.columns.tolist()

cols.remove('days_stable_min')
cols.insert(9, 'days_stable_min')

df = df[cols]
df.columns[:20]
```

```python
##--- adding drug loading column ---##
wells = [
    (df['well'].isin([0,8,16,24,32])), #80
    (df['well'].isin([1,9,17,25,33])),#70
    (df['well'].isin([2,10,18,26,34])),#60
    (df['well'].isin([3,11,19,27,35])),#50
    (df['well'].isin([4,12,20,28,36])),#40
    (df['well'].isin([5,13,21,29,37])),#30
    (df['well'].isin([6,14,22,30,38])),#20
    (df['well'].isin([7,15,23,31,39])),#10
    (df['well'].isin(np.arange(40,48)))#100
]

loadings = [80,70,60,50,40,30,20,10,100]

df['Drug_loadings_(wt%)'] = np.select(wells, loadings, default=None)

cols = df.columns.tolist()

cols.remove('Drug_loadings_(wt%)')
cols.insert(10, 'Drug_loadings_(wt%)')

df = df[cols]
df.columns[:20]
```

```python
df.to_csv(r'/home/lero/idrive/cmac/DDMAP/stability_studies_master/stability_dataset.csv')
```
