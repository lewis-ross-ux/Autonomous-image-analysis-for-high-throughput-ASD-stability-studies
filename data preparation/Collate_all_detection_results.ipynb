```python
import os
import pandas as pd
```

```python
file_path=r'/home/lero/idrive/cmac/DDMAP/stability_studies_master'

folders = ['40_C_75_RH', 
           '30_C_30_RH',
           '40_C_0_RH'
           ]
```

```python
stability_list = []

for folder in folders:
    folder_path = os.path.join(file_path, folder)
    
    for API_subfolder in os.listdir(folder_path):
        API_path = os.path.join(folder_path, API_subfolder, 'new_results')
        Stability_file = os.path.join(API_path, 'stability_summary.csv')
        if not os.path.exists(Stability_file):
            print(f"⚠️ Missing file: {Stability_file}")  
            continue  
        
        if os.path.exists(Stability_file):
            df = pd.read_csv(Stability_file)
            df['API'] = API_subfolder
            df['condition'] = folder
            stability_list.append(df)
            
            print(f'processed {Stability_file}')
        
            
            
API_stability_data = pd.concat(stability_list, axis=0, ignore_index=True)
Output_file=os.path.join(file_path, 'API_stability_APR_26.csv')
API_stability_data.to_csv(Output_file, index=False)
```
